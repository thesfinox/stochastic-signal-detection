```python
df = df.sort_values(by=['beta'])
```

```python
beta = df['beta']
```

```python
dim_mu_4_bar_end = df['dim_mu_4_bar'].apply(lambda x: json.loads(x)[-1])
dim_mu_6_bar_end = df['dim_mu_6_bar'].apply(lambda x: json.loads(x)[-1])
dim_mu_8_bar_end = df['dim_mu_8_bar'].apply(lambda x: json.loads(x)[-1])
```

```python
fig, ax = plt.subplots()
ax.plot(beta, dim_mu_4_bar_end, label=r'$-\mathrm{dim}\, \overline{\mu}_4$')
ax.plot(beta, dim_mu_6_bar_end, label=r'$-\mathrm{dim}\, \overline{\mu}_6$')
ax.plot(beta, dim_mu_8_bar_end, label=r'$-\mathrm{dim}\, \overline{\mu}_8$')
ax.set_xlabel(r'$\beta$')
ax.set_ylabel(r'$-\mathrm{dim}\, \overline{\mu}_n$')
ax.hlines(0, 0, 1.5, linestyle='--', colors='k')
ax.legend()
plt.tight_layout()
plt.show()
plt.close(fig)
```

```python
Up_start = df['Up_start'].apply(lambda x: json.loads(x)[-1])
Up_end = df['Up_end'].apply(lambda x: json.loads(x)[-1])
```

```python
fig, ax = plt.subplots()
ax.plot(beta, Up_start, label=r'$U^{\prime}[0]$')
ax.plot(beta, Up_end, label=r'$U^{\prime}[1]$')
ax.set_xlabel(r'$\beta$')
ax.set_ylabel(r'$U^{\prime}$')
ax.legend()
plt.tight_layout()
plt.show()
plt.close(fig)
```
